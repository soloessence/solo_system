```python
date = '2021-07-16'

sql = '''

select distinct code from ak_stock_info 
where code not in(select distinct code from ak_stock_daily where date='{}' )
--where code not like 'sz3%' 
--and code not in(select distinct code from ak_stock_daily)

'''.format(date)
stocks = pd.read_sql(sql, engine)
stocks.head()
```

```python
## 获取日K 数据  
#for code in stock_zh_a_spot_df['code']
for code in stocks['code']:
    print(code)
    df = ak.stock_zh_a_daily(symbol=code, start_date=date, end_date=date, adjust="qfq")
    
    #df = df.sort_values(by='date', ascending=True)
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date')
    
    df['code'] = code 
    
    df.to_sql("ak_stock_daily", engine, if_exists='append')
    
```
