```python
## ak_industry  -- only need run once 

ak_industry = ak.stock_board_industry_name_ths()
#ak_industry.to_sql("ak_industry", engine, if_exists='append', index=False)
```

```python
## 批量获取板块数据

date = '2021-07-09'

for ind in ak_industry['name']:
    df = ak.stock_board_industry_cons_ths(symbol=ind)
    df = df[['代码','名称','现价','涨速','换手',"量比","振幅",'流通股','流通市值','市盈率']]
    df.columns = ['code', 'name', 'price', 'speed','turnover',"qr",'vol','volume','equity', 'pe']
    df['board'] = ind
    df['date'] = date
    #df.replace('--',np.nan)
    #df = df.dropna()
    df['volume'] = df['volume'].apply(lambda x: x.replace('亿',''))
    df['equity'] = df['equity'].apply(lambda x: x.replace('亿',''))
    df = df.drop(df[(df.code == '暂无成份股数据')].index)

    df['price'] = df['price'].apply(lambda x: np.nan if x=='--' else x)
    df['speed'] = df['speed'].apply(lambda x: np.nan if x=='--' else x)
    df['turnover'] = df['turnover'].apply(lambda x: np.nan if x=='--' else x)
    df['qr'] = df['qr'].apply(lambda x: np.nan if x=='--' else x)
    df['vol'] = df['vol'].apply(lambda x: np.nan if x=='--' else x)
    df['volume'] = df['volume'].apply(lambda x: np.nan if x=='--' else x)
    df['equity'] = df['equity'].apply(lambda x: np.nan if x=='--' else x)
    df['pe'] = df['pe'].apply(lambda x: np.nan if x=='--' else x)
    
    df.to_sql("ak_industry_details_new", engine, index=False, if_exists='append')
```
