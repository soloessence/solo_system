```python
##以股票自身增长为Benchmark， 方便用来做策略与不采取策略的对比

import datetime
import backtrader as bt

                    
class EmaCross10And20(bt.SignalStrategy):

    params = dict(ema1=10, ema2=20)

    def notify_order(self, order):
        pass

    def notify_trade(self, trade):
        pass
#         if trade.isclosed:
#             print('profit {}'.format(trade.pnlcomm))

    def __init__(self):
        ema1 = bt.ind.EMA(period=self.params.ema1)
        ema2 = bt.ind.EMA(period=self.params.ema2)
        crossover = bt.ind.CrossOver(ema1, ema2)

        self.signal_add(bt.SIGNAL_LONG, crossover)
        

TIMEFRAMES = {
    None: None,
    'days': bt.TimeFrame.Days,
    'weeks': bt.TimeFrame.Weeks,
    'months': bt.TimeFrame.Months,
    'years': bt.TimeFrame.Years,
    'notimeframe': bt.TimeFrame.NoTimeFrame,
}


cerebro = bt.Cerebro()
cerebro.broker.set_cash(100000)
cerebro.addsizer(bt.sizers.AllInSizerInt) 

stock = 'sz002241'

df = ak.stock_zh_a_daily(symbol=stock, adjust="qfq", start_date='20160501', end_date='20210716').iloc[:, :6]

df.columns = [
    'datetime',
    'open',
    'close',
    'high',
    'low',
    'volume',
]

df.index = pd.to_datetime(df['datetime'])

data0 = bt.feeds.PandasData(
    dataname=df,
    fromdate=datetime.datetime.strptime('2016-05-24', '%Y-%m-%d'),
    todate=datetime.datetime.strptime('2021-07-16', '%Y-%m-%d'))

cerebro.adddata(data0, name='Data0')

cerebro.addstrategy(EmaCross10And20)


benchdata = data0
cerebro.addobserver(bt.observers.Benchmark,
                    data=benchdata,
                    timeframe=TIMEFRAMES['notimeframe'])

cerebro.run()
cerebro.plot()
```
