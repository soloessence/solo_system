```python
sql = '''

select code from ak_index_info 
where  1=1 

'''
index = pd.read_sql(sql, engine)
```

```python


from talib import abstract

for code in index['code']:
    print(code)
    df = ak.stock_zh_index_daily(symbol=code)
    
    #pro.daily(ts_code=code, start_date='20190524', end_date='20210611')
    
    df = df.reset_index()
    df = df.sort_values(by='date', ascending=True)
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date')
    
    macd  = abstract.MACD
    ma  = abstract.SMA
    ema = abstract.EMA 
    atr = abstract.ATR 
    
    macd_value, macdsignal, macdhist = macd(df['close'], fastperiod=12, slowperiod=26, signalperiod=9)

    ma_5 = ma(df, timeperiod=5)  # calculate on close prices by default
    ma_20 = ma(df, timeperiod=20)
    ma_60 = ma(df, timeperiod=60)
    ma_120 = ma(df, timeperiod=120)
    ema_20 = ema(df, timeperiod=20)
    ema_60 = ema(df, timeperiod=60)
    ema_120 = ema(df, timeperiod=120)
    atr_20 = atr(df, timeperiod=20 )
    
    df['MA5'] = ma_5
    df['MA20'] = ma_20
    df['MA60'] = ma_60
    df['MA120'] = ma_120
    df['MACD'] = macd_value
    df['EMA20'] = ema_20
    df['ATR20'] = atr_20
    df['code'] = code 
    
    df.to_sql("ak_index_daily", engine, if_exists='append')
```
