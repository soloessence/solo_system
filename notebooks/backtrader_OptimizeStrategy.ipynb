```python
import datetime 


##Jupyter 中居然不能用多线程!!!, 只能设maxcpus =1 
cerebro = bt.Cerebro(maxcpus=1,optreturn=True,runonce=True)
                     #runonce=not args.no_runonce,
                     #exactbars=args.exactbars,
                     #optdatas=not args.no_optdatas,
                     #optreturn=not args.no_optreturn)

start_cash = 100000
cerebro.broker.set_cash(start_cash)
cerebro.addsizer(bt.sizers.AllInSizerInt) 
    


stock = 'sz002241'

df = pd.read_sql('''select date as datetime, open, close, high, low, volume 
            from ak_stock_daily where code = '{}' and date<='2021-07-09' order by date asc  '''.format(stock), engine)


# 处理字段命名，以符合 Backtrader 的要求
df.columns = [
    'datetime',
    'open',
    'close',
    'high',
    'low',
    'volume',
]

df.index = pd.to_datetime(df['datetime'])


data0 = bt.feeds.PandasData(
    dataname=df,
    fromdate=datetime.datetime.strptime('2018-05-24', '%Y-%m-%d'),
    todate=datetime.datetime.strptime('2021-07-09', '%Y-%m-%d')
)


    
    
    
# Add the Data Feed to Cerebro
cerebro.adddata(data0)


#Add a strategy
cerebro.optstrategy(
    St,
    period=range(8, 21)
)

cerebro.addanalyzer(bt.analyzers.AnnualReturn, _name='_AnnualReturn')
cerebro.addanalyzer(bt.analyzers.Returns, _name='_Returns', tann=252)

#cerebro.addstrategy(St)
# Run over everything
stratruns = cerebro.run()


for stratrun in stratruns:
    for strat in stratrun:
        result = "{} --> 年化收益率: {:.2f}%".format(strat.params.period, strat.analyzers._Returns.get_analysis()['rnorm100'])
        print(result)


```
